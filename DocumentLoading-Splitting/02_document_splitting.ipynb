## MarkDown Text Splitter (Missing Part Exist)

```python
from langchain.text_splitter import MarkdownHeaderTextSplitter
```

```python
markdown_doc = """
# Whale Life \n \
## Introduction\n \
Whales are magnificent creatures that inhabit the oceans of the world. They are known for their immense size and graceful movements in the water.\n \
## Types of Whales\n \
There are several different types of whales, each with its own unique characteristics. Some common types of whales include:\n \
- Blue Whale\n \
- Humpback Whale\n \
- Killer Whale,\n \
## Whale Behavior\n \
Whales are highly intelligent and social animals. They often travel in pods and communicate with each other through a series of clicks, whistles, and songs.\n \
## Whale Conservation\n \
Due to factors such as pollution, climate change, and hunting, many whale species are endangered. Conservation efforts are crucial to protect these magnificent creatures and their habitats.\n \
## Conclusion\n \
Whales play a vital role in maintaining the balance of marine ecosystems. By learning more about them and taking steps to protect their environment, we can ensure a brighter future for these incredible creatures.
"""
```

```python
headers_to_split = [("# ", "Header 1"), ("## ","Header 2")]
```

```python
markdown_splitter = MarkdownHeaderTextSplitter(headers_to_split_on = headers_to_split)
mdsplilts = markdown_splitter.split_text(markdown_doc)
```
